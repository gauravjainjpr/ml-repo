```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
housing_dataset = pd.read_csv('./dataset/housing.csv')

df = housing_dataset

df.sample(5)
```

```python
# drop NAs

df = df.dropna()

df.shape
```

```python
df_new = df.drop(df.loc[df['median_house_value']==500001].index)

df_new = pd.get_dummies(df_new, columns = ['ocean_proximity'])
```

```python
df_new.sample(5)
```

```python
median = df_new['median_house_value'].median()

median
```

```python
df_new['above_median'] = (df_new['median_house_value'] - median) > 0
```

```python
df_new.sample(5)
```

```python
X = df_new.drop(['median_house_value','above_median'], axis=1)
Y = df_new['above_median']
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.20)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
      
logistic_model = LogisticRegression(solver='liblinear').fit(x_train, y_train)
```

```python
print('Training Score: ', logistic_model.score(x_train, y_train))
```

```python
y_pred = logistic_model.predict(x_test)
```

```python
df_pred_actual = pd.DataFrame({'predicted': y_pred, 'actual': y_test})

df_pred_actual.head(10)
```

```python
from sklearn.metrics import accuracy_score

print("Testing Score: ", accuracy_score(y_test, y_pred))
```
